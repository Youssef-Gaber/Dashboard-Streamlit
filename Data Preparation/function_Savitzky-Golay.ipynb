```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as ss
df=pd.read_csv("synchronous machine.csv",decimal=',',delimiter=';')

excitation_current= df['If']
load_current= df['Iy']
power_factor= df['PF']
power_factor_error= df['e']
excitation_changing= df['dIf']
```

```python
def savitzky_golay(data,filter_lenght, order):
    """data: function that will be filtered
        filter_lenght: lenght of the window
        order: order of the polynomial to fit the funtion"""
    unfiltered = data
    sg_filter = ss.savgol_filter(unfiltered, filter_lenght, order)
    plt.plot(unfiltered)
    plt.plot(sg_filter)
    return
```

```python
savitzky_golay(excitation_current,5,3)
```
