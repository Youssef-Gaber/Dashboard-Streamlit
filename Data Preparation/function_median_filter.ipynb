```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as ss
df=pd.read_csv("synchronous machine.csv",decimal=',',delimiter=';')

excitation_current= df['If']
load_current= df['Iy']
power_factor= df['PF']
power_factor_error= df['e']
excitation_changing= df['dIf']
```

```python
def median_filter(filter_lenght, data):
    """data: function that will be filtered
        filter_lenght: lenght of the window """
    from scipy.signal import medfilt
    medfilt_tair = medfilt(data, filter_lenght )
    plt.plot(data)
    plt.plot(medfilt_tair)
    return
```
