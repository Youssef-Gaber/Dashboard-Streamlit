```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as ss
df=pd.read_csv("synchronous machine.csv",decimal=',',delimiter=';')
```

```python
excitation_current= df['If']
load_current= df['Iy']
power_factor= df['PF']
power_factor_error= df['e']
excitation_changing= df['dIf']
```

```python
def moving_average (filter_length, data):
    tair_moving_average = np.convolve(data, np.ones((filter_length)), mode ="same")
    tair_moving_average /= filter_length
    tair_moving_average= tair_moving_average[round(filter_length/2): round(len(data)-filter_length/2)]
    original_function= data
    original_function= original_function[round(filter_length/2): round(len(data)-filter_length/2)] 
    plt.plot(original_function)
    plt.plot(tair_moving_average)
    plt.plot(tair_moving_average) 
    return 
```
